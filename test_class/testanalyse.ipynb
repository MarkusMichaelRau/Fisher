```python
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

Plot the dndz distributions
-------------------

```python
element1_pdf = np.loadtxt('element1_pdf.dat')
element2_pdf = np.loadtxt('element2_pdf.dat')
```

```python
plt.plot(element1_pdf[:, 0], element1_pdf[:, 1], '-', color='red', lw=2)
plt.plot(element2_pdf[:, 0], element2_pdf[:, 2], '--', color='black', lw=2)
plt.plot(element1_pdf[:, 0], element1_pdf[:, 2], '-', color='brown', lw=2)
plt.plot(element2_pdf[:, 0], element2_pdf[:, 1], '--', color='cyan', lw=2)
plt.xlabel('Redshift', fontsize=16)
plt.ylabel('Density', fontsize=16)
```

```python
element1_cl = np.loadtxt('element1cl.dat')
element2_cl = np.loadtxt('element2cl.dat')

```

```python
plt.plot(element1_cl[:, 0], (element1_cl[:, 3] - element2_cl[:, 1])/element1_cl[:, 3], '-')
plt.xlabel(r'$\ell$', fontsize=14)
plt.ylabel(r'$\Delta C_{\ell}$', fontsize=14)
plt.xscale('log')
plt.yscale('log')
```

```python
plt.plot(element1_cl[:, 0], (element1_cl[:, 1] - element2_cl[:, 3])/element1_cl[:, 1], '-')
plt.xlabel(r'$\ell$', fontsize=14)
plt.ylabel(r'$\Delta C_{\ell}$', fontsize=14)
plt.xscale('log')
plt.yscale('log')
```
