# Comparing Binning Methods
This notebook shows how the Kitching Figure of Merit changes when we change the number of bins for both equal number and equal size bins

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

# Auto Correlation Only
## Load the data

```python
sn_ratios = np.zeros((5, 3))
sn_ratios[:, 0] = list(range(1, 6))
# load in equal space
for i in range(4):
    sn_ratios[i, 1] = np.loadtxt("../out_w0_sn_all/equal_space/%d_bins/lensing_bins_%d_delta_1.02485379e-01/w0_sn_ratio.dat"%(i+1,i+1))
sn_ratios[4, 1] = np.loadtxt("../out_w0_sn_all/equal_space/5_bins/lensing_bins_5_delta_1.10870244e-01/w0_sn_ratio.dat")
#load in equal number
for i in range(5):
    sn_ratios[i, 2] = np.loadtxt("../out_w0_sn_all/equal_num/%d_bins/lensing_bins_%d_delta_1.02485379e-01/w0_sn_ratio.dat"%(i+1,i+1))
```

```python
df = pd.DataFrame(data=sn_ratios, columns=["Number of Bins", "Equal Space FoM", "Equal Number FoM"])\
       .set_index("Number of Bins")\
       .apply(lambda x: -x)
```

## Plots

```python
df
```

Plot of the raw signal to noise ratios (they are negative as per the kitching paper)

```python
df.plot(style='-o', figsize=(9,6))
plt.ylabel("Kitching FoM")
plt.show()
```

Now let's plot the signal to noise ratios as they appear in the Kitching paper, where they are normalized by the equal space signal to noise ratio at each number-of-bins

```python
df2 = df.copy()
```

```python
df2["Equal Space FoM"] = df2["Equal Space FoM"]/df["Equal Space FoM"]
df2["Equal Number FoM"] = df2["Equal Number FoM"]/df["Equal Space FoM"]
```

```python
df2.plot(style='-o', figsize=(9,6))
plt.ylabel("$F/F_{eqz}$")
plt.show()
```
